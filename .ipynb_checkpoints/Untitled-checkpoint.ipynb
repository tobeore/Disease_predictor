# Data Cleaning

```python
data.info()
```

The dataset used has 42 features consisting of 10 features with data type float64, 20 features with int64 and 12 features with object data type

```python
missing_values = data.isnull().sum()
missing_values_percentage = (missing_values/len(data))* 100
print(f'Table of Missing Values : \n{missing_values}')
print(f'Missing Value Percentage: \n{missing_values_percentage}')
```

```python
# ceating a table to display features that have missing values
Missing_vals = missing_values_percentage[missing_values_percentage >= 1 ]
Missing_vals = Missing_vals.reset_index().rename( columns = {0: 'count'})
Missing_vals
```

```python
plt.figure(figsize = (10,10))
sns.barplot(data = Missing_vals,  x = 'index', y = 'count')
plt.xticks(rotation = 90)
plt.show()
```

OBSERVATION:
1. Only CS_RACA and CS_ESCOL_N have percentage of missing values lower than 20% 
2. CRITERIO and EVOLUCAO both having missing values with percentages over 51%
3. COMUNIF, COPAISNF, COUFINF and TPAUTOCTO have missing values with the percentage of over 66%
4. DOENCA_TRA has a percentage missing value of 76.9%
5. While DT_OBITO, FLXRECEBI and NDUPLIC_N missing values with percentage above 99%

SOLUTION:
To free the data from missing values, we drop columns with their percentage of missing values 50% and replace columns with missing values percentage of 50% and below with their mean. This is because the mean is not affected by outliers

```python
# Define the threshold for dropping columns
drop_threshold = 50  

# Iterate over the columns and decide whether to drop or fill NaN with mean
for column in data.columns:
    missing_percentage = (data[column].isnull().sum() / len(data)) * 100

    if missing_percentage > drop_threshold:
        data.drop(columns=[column], inplace=True)
    else:
        # Handle based on the data type
        if data[column].dtype == 'object':
            # For object (categorical) columns, fill NaN with mode (most frequent value)
            data[column].fillna(data[column].mode()[0], inplace=True)
        else:
            # For numerical columns, fill NaN with mean
            data[column].fillna(data[column].mean(), inplace=True)


# Verify the changes
print(data.isnull().sum())  # To check if there are any remaining NaN values
```
